```python
import pandas as pd
import requests
import datetime
from datetime import datetime
```

```python
# Inizializza la lista data_list
data_list = []

# Effettua 31 chiamate all'API di Alternative.me
for i in range(365):
  # Prepara la richiesta HTTP
  url = "https://api.alternative.me/fng/?limit=365&date_format=us" + str(i)
  headers = {'Content-Type': 'application/json'}

  # Effettua la richiesta HTTP
  response = requests.get(url, headers=headers)

  # Carica i dati in formato JSON
  data = response.json()

  # Estrai il valore "value" e il timestamp
  value = data["data"][i]["value"]
  timestamp = data["data"][i]["timestamp"]

  # Aggiungi i valori "value" e "timestamp" nella lista data_list
  data_list.append((value, timestamp))
```

```python
#converto la lista in dataframe e trasformo le date in datetime
df_greeded = pd.DataFrame(data_list, columns=['greed_and_fear', 'time'])
#df_greeded["time"] = pd.to_datetime(df_greeded["time"], unit='s')

# Formatta la colonna "timestamps" come YYYY-MM-DD
#df_greeded["time"] = df_greeded["time"].dt.strftime("%Y-%m-%d")
df_greeded.head()
```

```python
#"dd-mm-yyyy" in una data nel formato "yyyy-mm-dd"
def convert_date(date_str):
    date = datetime.strptime(date_str, "%d-%m-%Y")
    return date.strftime("%Y-%m-%d")

    # Usa il metodo .apply() sulla colonna "time" del DataFrame, passandogli la funzione convert_date come argomento
df_greeded["time"] = df_greeded["time"].apply(convert_date)

df_greeded.head()
```
